#1.Linear Regression

##1.1.Linear Regression from Scratch

```python
# a sample
x = 2
y = 18
```

```python
# init weights
w = 2
b = 2
```

```python
# forward
def predict(x, w, b):
    return x*w + b

y_hat = predict(x, w, b)
y_hat
```

```python
# compute loss
def compute_loss(y_hat, y):
    return (y_hat-y)*(y_hat-y)

loss = compute_loss(y_hat, y)
loss
```

```python
# compute gradient
def compute_gradient(x, y, y_hat):
    dw = 2*x*(y_hat-y)
    db = 2*(y_hat-y)
    return dw, db

dw, db = compute_gradient(x, y, y_hat)
dw, db
```

```python
# update weights
lr = 0.1

def update_weights(w, b, dw, db, lr):
    new_w = w - lr*dw
    new_b = b - lr*db
    return new_w, new_b

new_w, new_b = update_weights(w, b, dw, db, lr)
new_w, new_b
```

```python
X = [2, 3]
Y = [18, 22]

w = 2
b = 2
lr = 0.1
num_epochs = 50

losses = []
for epoch in range(num_epochs):
    epoch_loss = []
    for x, y in zip(X, Y):
        # compute y_hat
        y_hat = predict(x, w, b)

        # comput loss
        loss = compute_loss(y_hat, y)
        epoch_loss.append(loss)

        # compute gradient
        dw, db = compute_gradient(x, y, y_hat)

        # update weights
        w, b = update_weights(w, b, dw, db, lr)
    losses.append(sum(epoch_loss)/len(epoch_loss))
```

```python
losses
```

```python
w, b
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
```

```python
Y_pred = [x*w + b for x in X]
plt.plot(X, Y_pred, 'r')
plt.scatter(X, Y)
```

```python
# init weights
theta = np.array([2, 2]) # num_features: x.shape[1]
theta
```

```python
# forward
def predict(x, theta):
    return x.dot(theta)

y_hat = predict(x, theta)
y_hat
```

```python
# compute loss
def compute_loss(y_hat, y):
    return (y_hat-y)*(y_hat-y)

loss = compute_loss(y_hat, y)
loss
```

```python
# compute gradient
def compute_gradient(x, y, y_hat):
    d_theta = 2*x*(y_hat-y)
    return d_theta

d_theta = compute_gradient(x, y, y_hat)
d_theta
```

```python
# update weights
lr = 0.1

def update_weights(theta, d_theta, lr):
    new_theta = theta - lr*d_theta
    return new_theta

new_theta = update_weights(theta, d_theta, lr)
new_theta
```

##2.2.Logistic Regression using Numpy for One Sample

```python
X = np.array([
    [1.0, 0.5],
    [2.0, 1.0],
    [3.0, 1.5],
    [2.0, 2.0],
    [1.0, 2.5],
    [2.0, 3.0],
    [1.0, 3.5],
    [2.0, 4.0],
])
Y = np.array([0, 0, 1, 0, 0, 1, 1, 1])
```

```python
X, Y
```

```python
n_samples = X.shape[0]
```

```python
# vector [1, X]
X = np.hstack([np.ones((n_samples, 1)), X])
X
```

```python
X.shape
```

```python
n_features = X.shape[-1]
theta = np.random.rand(n_features)
theta
```

```python
def logistic_function(x):
  return 1/ (1 + np.exp(-x))

def predict(x, theta):
  z = np.dot(x , theta)
  y_pred = logistic_function(z)
  return y_pred

def compute_cost(y_pred, y):
  cost = -1*((y * np.log(y_pred)) + ((1 - y) * np.log(1 - y_pred)))
  return cost

def compute_gradient(x, y, y_pred):
  gradient = np.dot(x.transpose(), (y_pred - y))
  return gradient

def update_weight(theta, lr, gradient):
  new_theta = theta - lr*gradient
  return new_theta
```

```python
# param
n_epochs = 10
lr = 0.001

losses = []
for epoch in range(n_epochs):
  epoch_loss = []
  for i in range(n_samples):
    # get a sample
    x = X[i,:]
    y = Y[i]

    # predict y_hat
    y_pred = predict(x, theta)

    # compute loss
    loss = compute_cost(y_pred, y)
    epoch_loss.append(loss)

    # compute gradient
    gradient = compute_gradient(x, y, y_pred)

    # update weights
    theta = update_weight(theta, lr, gradient)
  losses.append(sum(epoch_loss)/len(epoch_loss))

theta, losses
```

```python
losses
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel('epoch')
plt.ylabel('losses')
plt.show()
```

```python
X_test = np.array([1.0, 2.0, 0.25])
predict(X_test, theta)
```

```python
X_test = np.array([1.0, 1.0, 4.5])
predict(X_test, theta)
```

##2.3.Logistic Regression for N Sample

```python
X = np.array([
    [1.0, 0.5],
    [2.0, 1.0],
    [3.0, 1.5],
    [2.0, 2.0],
    [1.0, 2.5],
    [2.0, 3.0],
    [1.0, 3.5],
    [2.0, 4.0],
])
Y = np.array([0, 0, 1, 0, 0, 1, 1, 1])
```

```python
n_samples = X.shape[0]
```

```python
# vector [1, X]
X = np.hstack([np.ones((n_samples, 1)), X])
X
```

```python
n_features = X.shape[-1]
theta = np.random.rand(n_features)
theta
```

```python
def logistic_function(x):
  return 1/ (1 + np.exp(-x))

def predict(x, theta):
  z = np.dot(x , theta)
  y_pred = logistic_function(z)
  return y_pred

def compute_cost(y_pred, y):
  m = len(y)
  error = (y * np.log(y_pred)) + ((1 - y) * np.log(1 - y_pred))
  cost = -1 / m * sum(error)
  return cost

def compute_gradient(x, y, y_pred):
  m = len(y)
  gradient = 1 / m * np.dot(x.transpose(), (y_pred - y))
  return gradient

def update_weight(theta, lr, gradient):
  new_theta = theta - lr*gradient
  return new_theta
```

```python
# param
n_epochs = 10
lr = 0.001

losses = []
for epoch in range(n_epochs):
  # get all the samples
  x = X
  y = Y

  # predict y_hat
  y_pred = predict(x, theta)

  # compute loss
  loss = compute_cost(y_pred, y)
  losses.append(loss)

  # compute gradient
  gradient = compute_gradient(x, y, y_pred)

  # update weights
  theta = update_weight(theta, lr, gradient)

theta, losses
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel('epoch')
plt.ylabel('losses')
plt.show()
```

```python
X_test = np.array([1.0, 2.0, 0.25])
predict(X_test, theta)
```

```python
X_test = np.array([1.0, 1.0, 4.5])
predict(X_test, theta)
```

```python
def logistic_function(x):
  return 1/ (1 + np.exp(-x))

def predict(x, theta):
  z = np.dot(x , theta)
  y_pred = logistic_function(z)
  return y_pred

def compute_cost(y_pred, y):
  m = len(y)
  error = (y * np.log(y_pred)) + ((1 - y) * np.log(1 - y_pred))
  cost = -1 / m * sum(error)
  return cost

def compute_gradient(x, y, y_pred):
  m = len(y)
  gradient = 1 / m * np.dot(x.transpose(), (y_pred - y))
  return gradient

def update_weight(theta, lr, gradient):
  new_theta = theta - lr*gradient
  return new_theta
```
